### Correlation in Numpy and Pandas

```python

import numpy as np

height = [60, 62, 65, 68, 70, 74]
weight = [140, 138, 150, 166, 190, 250]

print(np.corrcoef(height, weight))
print('') #  line break


# print only the r rather than the whole matrix; round to 2 decimals
print(round(np.corrcoef(height, weight)[0][1], 2))
```

```python
import pandas as pd
df = pd.read_csv("http://www.ishelp.info/data/insurance.csv")

df.corr()
```

```python
df.charges.corr(df.age)
```

### Correlation in SciPy


```python
from scipy import stats

corr = stats.pearsonr(df.charges, df.age)
corr
```

```python
# if they have same format

print('r: \t' + str(round(corr[0], 4)))
print('p-value:' + str(round(corr[1], 4)))
```

### Assumptions
1. continous data
2. linear relation
3. homoskedastic relation
---
##### definitions
    # for non continous - the values b/w intervals or hidden are equal as the given rank would not be true
    # linear relation - where relation can be represented via a straight line if not make the linee straight
    # homoskedastic relation - consistency(spread) across  all values of x & y
#####  resolutions - kendall's(tau) or spearman(rho)


```python
# P value - prob that results wont be repeated in next data collection OR
# likelihood that the resulyts are not representative of overall population

# low p = low likelihood to repeat
# hight p value = high likelihood



# Create an empty DataFrame to store the correlations and p-values
corr_df = pd.DataFrame(columns=['r', 'p-value'])

for col in df:
  if pd.api.types.is_numeric_dtype(df[col]):      # Only calculate r, p-value for the numeric columns
    r, p = stats.pearsonr(df.charges, df[col])    # .pearsonr() returns two values in a list, store them individually
    corr_df.loc[col] = [round(r, 3), round(p, 3)] # Add the r & p for this col into the corr_df

corr_df.sort_values(by=['r'], ascending=False)    # Sort and display the corr_df
```

```python
def bivariate_stats(df, label):
  corr_df = pd.DataFrame(columns=['r', 'p-value'])

  for col in df:
    if pd.api.types.is_numeric_dtype(df[col]) and col != label:
      r, p = stats.pearsonr(df[label], df[col])
      corr_df.loc[col] = [round(r, 3), round(p, 3)]

  return corr_df.sort_values(by=['r'], ascending=False)
```
